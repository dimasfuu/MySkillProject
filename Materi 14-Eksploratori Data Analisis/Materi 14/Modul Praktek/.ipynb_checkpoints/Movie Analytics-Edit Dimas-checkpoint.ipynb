# 3. Data Cleansing

Agar data yang disajikan bisa lebih lengkap ada beberapa opsi yang dilakukan ketika menemukan fenomena adanya data yang tidak terisi lengkap sehingga menyebabkan sebaran atau distribusi data menjadi lebih sulit untuk dilakukan analisis. Opsi tersebut adalah : 
1. Melakukan interpolasi dengan pengisian/mengganti data missing value dengan statistik tertentu(rata-rata, median atau modus)
2. Membuang/menghapus data tersebut

mengingat banyaknya data **Missing value** di beberapa kolom/fitur maka opsi yang diambil adalah dengan melakukan interpolasi data.

```python
dc = movies_data.copy()
```

```python
dc.info()
```

```python
dc.isnull().sum()
```

```python
# Melakukan pengisian untuk kolom "belongs_to_collection"
dc['belongs_to_collection'].fillna("{id:"+dc["id"].astype(str)+",name:"+dc["original_title"],inplace=True)
dc.isnull().sum()
```

```python
# Melakukan pengisian kolom untuk "Homepage"
dc["homepage"].fillna("Website Not Available",inplace=True)
dc.isnull().sum()
```

```python
dc.head()
```

```python
# Melakukan pengisian untuk kolom "imdb_id"
dc["imdb_id"].fillna("imdb_id Not Available",inplace=True)
dc.isna().sum()
```

```python
# Melakukan pengisian untuk kolom "original_language"
dc["original_language"].fillna(dc["original_language"].value_counts().max(), inplace=True)
dc.isna().sum()
```

```python
dc.head(5)
```

```python
# Melakukan pengisian untuk kolom "overview"
dc["overview"].fillna("(Overview Not Available)", inplace=True)
dc.isna().sum()
```

```python
dc[["popularity","poster_path","production_companies","production_countries"]].head()
```

```python
# Melakukan pengisian kolom/fitur untuk popularity, poster_path, production_companies and countries
dc["popularity"].fillna(dc["popularity"].mean(),inplace=True)
dc["poster_path"].fillna("(poster doesn't exist)",inplace=True)
dc["production_companies"].fillna("Not Available",inplace=True)
dc["production_countries"].fillna("Not Available",inplace=True)
```

```python
dc.isna().sum()
```

```python
dc.head(2)
```

```python
movies_data.info()
```

```python
# Melakukan pengisian missing value untuk kolom "release_date,revenue,runtime,spoken_languages,status,tagline"
dc["release_date"].fillna(dc["release_date"].mode(),inplace=True)
dc["revenue"].fillna(dc["revenue"].mean(),inplace=True)
dc["runtime"].fillna(dc["runtime"].mean(),inplace=True)
dc["spoken_languages"].fillna(dc["spoken_languages"].mode(),inplace=True)
dc["status"].fillna(dc["status"].mode(),inplace=True)
dc["tagline"].fillna("Tagline not available")
```

```python
def select_columns(df: pd.DataFrame, features: list,  target: list = "rating"):

  df = df.dropna().reset_index(drop = True)
  X = df[features]
  y = (df[target] >= 3).astype("int")

  return X, y
```

```python
from sklearn.model_selection import train_test_split
```

```python
features = ["budget", "popularity", "vote_average"]
X, y = select_columns(new_df, features = features)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
minmax = MinMaxScaler()
```

```python
minmax.fit(X_train)
```

```python
def scaling(df, scaler):

  # df["runtime	"] = df["runtime"].div(60)

  scaled_array = scaler.transform(df)
  scaled_df = pd.DataFrame(scaled_array, columns = scaler.get_feature_names_out())

  return scaled_df
```

```python
X_train_scaled = scaling(X_train, minmax)
X_test_scaled = scaling(X_test, minmax)
```

```python
X_test_scaled.iloc[0]
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors = 5)
dt = DecisionTreeClassifier()
```

```python
knn.fit(X_train_scaled, y_train)
dt.fit(X_train_scaled, y_train)
```

```python
y_train_pred_knn = knn.predict(X_train_scaled)
y_train_pred_dt = dt.predict(X_train_scaled)
```

```python
y_test_pred_knn = knn.predict(X_test_scaled)
y_test_pred_dt = dt.predict(X_test_scaled)
```

```python
pd.crosstab(y_test, y_test_pred_knn)
```

```python
pd.crosstab(y_test, y_test_pred_dt)
```
