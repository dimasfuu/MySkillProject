Agar data yang disajikan bisa lebih lengkap ada beberapa opsi yang dilakukan ketika menemukan fenomena adanya data yang tidak terisi lengkap sehingga menyebabkan sebaran atau distribusi data menjadi lebih sulit untuk dilakukan analisis. Opsi tersebut adalah : 
1. Melakukan interpolasi dengan pengisian/mengganti data missing value dengan statistik tertentu(rata-rata, median atau modus)
2. Membuang/menghapus data tersebut

mengingat banyaknya data **Missing value** di beberapa kolom/fitur maka opsi yang diambil adalah dengan melakukan interpolasi data.

Setelah kita melakukan cleansing data, maka selanjutnya pada tahapan ini kita akan melihat bagaimana kondisi data dari sisi ukuran pemusatan, penyebaran data lalu hubungan antar variabel yang timbul antara variabel numerik

## 4.1. Statistik Deskriptif

```python
dc.describe()
```

```python
# Memasukan variabel tambahan "profit" dengan menghitung selish antara "revenue" dengan "budget"
dc["profit"] = dc["revenue"] - dc["budget"]
dc.describe()
```

```python
# Melihat sebaran data secara histogram
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

for col in dc.select_dtypes(include=['number']).columns:  # Filter kolom numerik saja
    # Plot histogram
    sns.histplot(data=dc, x=col, bins=30, color='blue', kde=False, stat="density")  # kde=False prevents automatic density curve
    
    # menghitung nilai rata-rata dan standar deviasi dari kolom
    mean = dc[col].mean()
    std = dc[col].std()
    
    # membuat nilai distribusi normal dari setiap kolom
    x = np.linspace(dc[col].min(), dc[col].max(), 30)
    y = norm.pdf(x, mean, std)
    
    # memploting kurva/garis distribusi normal
    plt.plot(x, y, color='red', label='Normal Distribution')
    
    plt.title(f"Histogram sebaran data {col}")
    plt.ylabel("Density")
    plt.xlabel(col)
    plt.legend()
    plt.show()
```

## 4.2. Melihat outlier dari boxplot

```python
for col in dc.select_dtypes(include=['number']).columns:  # Filter kolom numerik saja
    # Plot boxplot
    sns.boxplot(data=dc, x=col, color='blue', ) 
    plt.title(f"Boxplot {col}")
    plt.ylabel("Density")
    plt.xlabel(col)
    plt.legend()
    plt.show()
```

Dari data boxplot diatas dapat terlihat bahwa masih banyak sekali data-data outlier yang terdapat pada data frame. Outlier dapat berguna dalam melihat apakah model yang akan kita bangun akan menjadi model yang Good fit, Underfit atau Overfit. Untuk permodelan dapat dilakukan dua kondisi, apakah dengan membuang data yang bersifat outlier atau tetap mempertahankan nilai yang outlier.

Untuk membedakan dataset asli tanpa mengalami reduksi data yang terkena outlier, digunakan dataset sebelumnya yaitu "dc" dan untuk data tanpa outlier akan menggunakan dataset "do".

```python
do = dc.copy()
do.sample()
```

```python
import pandas as pd
for col in do.select_dtypes(include=["number"]).columns:
    # Asumsikan "do" adalah dataset Anda
    Q1 = do[col].quantile(0.25)
    Q3 = do[col].quantile(0.75)
    IQR = Q3 - Q1
    
    # Tentukan batas
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Filter data tanpa outlier
    do_clean = do[(do[col] >= lower_bound) & (do[col] <= upper_bound)]
    
    print(do_clean)
```

```python
do_clean.describe()
```

Seperti yang terlihat pada hasil pengolahan data tanpa outlier, bahwa data yang tersisa dari jumlah data awal sejumlah 45463 data menjadi hanya 15205 data saja, jumlah data yang tersisa hanya sekitar dibawah 33% dari awal. dan data tersebut akan sangat menentukan terhadap hasil dari uji dan latih dari machine learning yang akan dibuat.

### Melihat Boxplot setelah dilakukan reduksi data outlier

```python
for col in do_clean.select_dtypes(include=['number']).columns:
    sns.boxplot(data=do_clean,x=col,color='green')
    plt.title(f"Boxplot setelah reduksi untuk kolom {col}")
    plt.show()
```

## 4.3. Melihat ukuran pemusatan data

Dalam statistik biasanya akan dilihat ukuran pemusatan data yang akan digunakan untuk melihat apakah data tersebut cenderung memiliki kecenderungan untuk terpusat atau tidak. ada 3 ukuran statistik yang biasa digunakan untuk melihat pemusatan data, diantaranya : 
1. Rata-rata
2. Median
3. Modus

```python
# Ukuran pemusatan data untuk data awal/ 'dc'
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

for col in dc.select_dtypes(include=["number"]).columns:
    # menghitung nilai mean median dan modus
    mean = dc[col].mean()
    median = dc[col].median()
    mode = dc[col].mode()[0] 
    
    # membuat dataframe baru
    pusat = pd.DataFrame({
        "Central": ["Mean", "Median", "Mode"],
        "Nilai": [mean, median, mode]
    })
    
    # Membuat barplot
    sns.barplot(data=pusat, x="Central", y="Nilai", hue="Central", color='green')
    plt.title(f"Ukuran Pemusatan Data untuk Kolom {col}")
    plt.ylabel("Nilai")
    plt.xlabel("Jenis Statistik")
    plt.show()
```

```python
# Ukuran pemusatan data untuk data setelah cleaning outlier/ 'do_clean'
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

for col in do_clean.select_dtypes(include=["number"]).columns:
    # menghitung nilai mean median dan modus
    mean = do_clean[col].mean()
    median = do_clean[col].median()
    mode = do_clean[col].mode()[0] 
    
    # membuat dataframe baru
    pusat1 = pd.DataFrame({
        "Central": ["Mean", "Median", "Mode"],
        "Nilai": [mean, median, mode]
    })
    
    # Membuat barplot
    sns.barplot(data=pusat1, x="Central", y="Nilai", hue="Central", color='green')
    plt.title(f"Ukuran Pemusatan Data untuk Kolom {col}")
    plt.ylabel("Nilai")
    plt.xlabel("Jenis Statistik")
    plt.show()
```

## 4.4. Ukuran sebaran data

Pada ukuran sebaran data, dapat dilihat seberapa besar variasi, kemencengan dan keruncingan data yang akan disajikan. Data variansi berguna untuk melihat seberapa besar data tersebut tersebar, sementara kemencengan/skewness dilihat untuk milihat apakah data tersebut simetris atau tidak (sebagai indikasi apakah data tersebut semakin mengikuti distribusi normal atau tidak) sementara keruncingan digunakan untuk mengetahui apakah terdapat nilai-nilai ekstrim.

```python
# Menggunakan data frame awal
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

def sebar(data):
    for col in data.select_dtypes(include=["number"]).columns:
        # menghitung nilai mean median dan modus
        std = data[col].std()
        skew = data[col].skew()
        kurt = data[col].kurt()
        
        # membuat dataframe baru
        sebar = pd.DataFrame({
            "disp": ["std", "skew", "kurt"],
            "Nilai": [std, skew, kurt]
        })
        
        # Membuat barplot
        sns.barplot(data=sebar, x="disp", y="Nilai", hue="disp", color='red')
        plt.title(f"Ukuran Sebaran Data untuk Kolom {col}")
        plt.ylabel("Nilai")
        plt.xlabel("Jenis Statistik")
        plt.show()
```

## 4.5. Scatterplot

Scatterplot digunakan untuk melihat sebaran data secara pencaran per individu data, selain itu scatterplot juga dapat digunakan untuk memproyeksikan seperti apa persamaan dari grafik permodelan yang akan terbentuk.

```python
# Define the function
def pencar(data):
    numeric_cols = data.select_dtypes(include=["number"]).columns  
    
    for i, col1 in enumerate(numeric_cols):  
        for col2 in numeric_cols[i+1:]:  
            sns.scatterplot(data=data, x=col1, y=col2, label='scatter', color='blue')
            plt.title(f"Scatter plot untuk {col1} vs {col2}")
            plt.legend()
            plt.show()
```

```python
def select_columns(df: pd.DataFrame, features: list,  target: list = "rating"):

  df = df.dropna().reset_index(drop = True)
  X = df[features]
  y = (df[target] >= 3).astype("int")

  return X, y
```

```python
from sklearn.model_selection import train_test_split
```

```python
features = ["budget", "popularity", "vote_average"]
X, y = select_columns(new_df, features = features)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
minmax = MinMaxScaler()
```

```python
minmax.fit(X_train)
```

```python
def scaling(df, scaler):

  # df["runtime	"] = df["runtime"].div(60)

  scaled_array = scaler.transform(df)
  scaled_df = pd.DataFrame(scaled_array, columns = scaler.get_feature_names_out())

  return scaled_df
```

```python
X_train_scaled = scaling(X_train, minmax)
X_test_scaled = scaling(X_test, minmax)
```

```python
X_test_scaled.iloc[0]
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors = 5)
dt = DecisionTreeClassifier()
```

```python
knn.fit(X_train_scaled, y_train)
dt.fit(X_train_scaled, y_train)
```

```python
y_train_pred_knn = knn.predict(X_train_scaled)
y_train_pred_dt = dt.predict(X_train_scaled)
```

```python
y_test_pred_knn = knn.predict(X_test_scaled)
y_test_pred_dt = dt.predict(X_test_scaled)
```

```python
pd.crosstab(y_test, y_test_pred_knn)
```

```python
pd.crosstab(y_test, y_test_pred_dt)
```
